```python
Lab 2.1: Number Classifier ○ Objective: Students will write a script to classify a user-provided integer as positive, negative, or zero, and also check if it's even or odd.

num = int(input("Enter a number: ") )
if num < 0:
    text = f"{num} is negative"
elif num > 0:
    text = f"{num} is positive"
else:
    text = f"{num} is zero"

if num == 0:
    print(f"{text}.")
elif num % 2 == 0:
    print(f"{text} and even number.")
else:
    print(f"{text} and odd number.")
20 is positive and even number.
```

```python
ACTIVITY 2.1 — อธิบายลำดับการทำงาน + Output
Block A: Evaluation Check

temp = 25
is_raining = True

if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.")
else:
    print("Unsure about the weather.")
```

# answer block A

ไล่ทีละเงื่อนไข (ค่า: temp = 25, is_raining = True)
บรรทัด if: temp > 20 and not is_raining

temp > 20 → 25 > 20 → True
not is_raining → not True → False
True and False → False  → ข้าม

บรรทัด elif: temp <= 20 or is_raining

temp <= 20 → 25 <= 20 → False
is_raining → True
False or True → True  → ทำงานบรรทัดนี้

Output ของ Block A: Stay indoors or bring an umbrella.
จุดสำคัญ: and ต้องจริงทั้งคู่ / or ขอแค่ข้างใดข้างหนึ่งจริง / not กลับค่า True↔False

```python
Block B: Vectorized Filter

phrase = "Python 2026 Engine"
consonants = {c.lower() for c in phrase if c.isalpha() and c.lower() not in 'aeiou'}
print(f"Unique consonants: {consonants}")
```

ACTIVITY 2.2 — ออกแบบเกม Choose Your Own Adventure
เนื้อเรื่อง: "หลงป่ายามค่ำคืน 
[เริ่มเรื่อง]
คุณตื่นขึ้นมากลางป่า มืดแล้ว มีทางแยกอยู่ตรงหน้า

จุดตัดสินใจ 1: input("ไปทางไหน? (left/right): ")
 ├── left  → เจอถ้ำ                    → ไปจุดตัดสินใจ 2
 ├── right → เจอแม่น้ำ                 → ไปจุดตัดสินใจ 3
 └── อื่น ๆ → "คุณหลงทางในความมืด GAME OVER"

จุดตัดสินใจ 2 (ถ้ำ): input("เข้าถ้ำไหม? (yes/no): ")
 ├── yes → เจอสมบัติ! "คุณรวยแล้ว 🏆 ENDING A: ผู้ชนะ"
 └── no  → "คุณนอนรอจนเช้าแล้วเดินออกมา ENDING B: รอดชีวิต"

จุดตัดสินใจ 3 (แม่น้ำ): input("ข้ามแม่น้ำไหม? (swim/wait): ")
 ├── swim → "กระแสน้ำแรงเกินไป GAME OVER"
 └── wait → "เรือผ่านมารับ ENDING C: ได้รับการช่วยเหลือ"

```python
choice1 = input("ไปทางไหน? (left/right): ")

if choice1 == "left":
    choice2 = input("เข้าถ้ำไหม? (yes/no): ")
    if choice2 == "yes":
        print("คุณรวยแล้ว 🏆 ENDING A: ผู้ชนะ")
    elif choice2 == "no":
        print("คุณนอนรอจนเช้าแล้วเดินออกมา ENDING B: รอดชีวิต")
    else:
        print("การตัดสินใจไม่ชัดเจน GAME OVER")
elif choice1 == "right":
    choice3 = input("ข้ามแม่น้ำไหม? (swim/wait): ")
    if choice3 == "swim":
        print("กระแสน้ำแรงเกินไป GAME OVER")
    elif choice3 == "wait":
        print("เรือผ่านมารับ ENDING C: ได้รับการช่วยเหลือ")
    else:
        print("การตัดสินใจไม่ชัดเจน GAME OVER")
else:
    print("คุณหลงทางในความมืด GAME OVER")
```
