```python
Lab 2.1: Number Classifier ○ Objective: Students will write a script to classify a user-provided integer as positive, negative, or zero, and also check if it's even or odd.

num = int(input("Enter a number: ") )
if num < 0:
    text = f"{num} is negative"
elif num > 0:
    text = f"{num} is positive"
else:
    text = f"{num} is zero"

if num == 0:
    print(f"{text}.")
elif num % 2 == 0:
    print(f"{text} and even number.")
else:
    print(f"{text} and odd number.")
20 is positive and even number.
```

```python
choice1 = input("ไปทางไหน? (left/right): ")

if choice1 == "left":
    choice2 = input("เข้าถ้ำไหม? (yes/no): ")
    if choice2 == "yes":
        print("คุณรวยแล้ว 🏆 ENDING A: ผู้ชนะ")
    elif choice2 == "no":
        print("คุณนอนรอจนเช้าแล้วเดินออกมา ENDING B: รอดชีวิต")
    else:
        print("การตัดสินใจไม่ชัดเจน GAME OVER")
elif choice1 == "right":
    choice3 = input("ข้ามแม่น้ำไหม? (swim/wait): ")
    if choice3 == "swim":
        print("กระแสน้ำแรงเกินไป GAME OVER")
    elif choice3 == "wait":
        print("เรือผ่านมารับ ENDING C: ได้รับการช่วยเหลือ")
    else:
        print("การตัดสินใจไม่ชัดเจน GAME OVER")
else:
    print("คุณหลงทางในความมืด GAME OVER")
```
