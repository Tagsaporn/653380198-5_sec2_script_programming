## Lab 3.1 — Part 1: Single Number Multiplication Table ตารางสูตรคูณเดียว

````python
# --- Lab 3.1 Part 1: Single Number Multiplication Table ---
n = int(input("Enter a number for its multiplication table (e.g., 7): "))
for i in range(1, 13):
    print(f"{n} x {i} = {n * i}")

print()  # spacer
````

## Lab 3.1 — Part 2: Full 12x12 Multiplication Grid ตารางสูตรคูณ12x12 (ลูปซ้อนลูป)

````python
# --- Lab 3.1 Part 2: Full 12x12 Multiplication Grid ---
print("12x12 Multiplication Table:\n")
for i in range(1, 13):        # ลูปนอก -> แถว
    for j in range(1, 13):    # ลูปใน -> คอลัมน์
        print(f"{i * j:4d}", end="")
    print()

print()  # spacer
````

## Lab 3.2 — Part 1: Countdown Timer นับถอยหลัง

````python
# --- Lab 3.2 Part 1: Countdown Timer ---
start = int(input("Enter starting number for countdown: "))
while start >= 0:
    print(f"Countdown: {start}")
    start -= 1

print("Blast off!")
````

## Lab 3.2 — Part 2: Simple Guessing Game เกมทายเลข

````python
# --- Lab 3.2 Part 2: Simple Guessing Game ---
secret_number = 42

while True:
    guess = int(input("Guess the number: "))

    if guess < secret_number:
        print(f"{guess} is too low! Try again.")
    elif guess > secret_number:
        print(f"{guess} is too high! Try again.")
    else:
        print("Congratulations! You guessed it!")
        break
````

### Lab 3.2 Part 2 — Challenge: limited number of attempts (using while/else)

````python
# --- Lab 3.2 Part 2 (Challenge): Guessing game with limited attempts ---
secret_number = 42
max_attempts = 5
attempts = 0

while attempts < max_attempts:
    guess = int(input(f"Guess the number (Attempt {attempts + 1}/{max_attempts}): "))
    attempts += 1

    if guess < secret_number:
        print(f"{guess} is too low! Try again.")
    elif guess > secret_number:
        print(f"{guess} is too high! Try again.")
    else:
        print("Congratulations! You guessed it!")
        break
else:
    # Runs only if the while loop completes WITHOUT hitting break
    print(f"Sorry, you ran out of attempts. The secret number was {secret_number}.")
````

### Snippet A: Infinite Loop
**Bug:** ตัวแปร `counter` ไม่เคยถูกอัปเดตค่าเลยในลูป ทำให้เงื่อนไข `counter < 5` เป็นจริงตลอดไป กลายเป็นลูปไม่รู้จบ

````python
# Buggy version (do NOT run as-is, it never ends):
# counter = 0
# while counter < 5:
#     print(counter)
#     # Missing counter increment!

# --- Fixed version ---
counter = 0
while counter < 5:
    print(counter)
    counter += 1  # FIX: update the loop control variable so the condition can become False
````

### Snippet B: Incorrect Range
**Bug:** ต้องการพิมพ์เลข 1 ถึง 5 แต่เขียน range(1, 5) ซึ่งจะพิมพ์ได้แค่ 1, 2, 3, 4 เพราะค่า stop ใน range() จะไม่ถูกนับรวม (exclusive)

````python
# Buggy version:
# for i in range(1, 5):  # Goal: print 1, 2, 3, 4, 5
#     print(i)

# --- Fixed version ---
for i in range(1, 6):  # FIX: stop must be 6 so that 5 is included
    print(i)
````

### Snippet C: Misplaced Break
**Bug:** ** โค้ดเช็คเงื่อนไขและสั่ง `break` ก่อน แล้วค่อยมีบรรทัด `print(char)` อยู่หลังจากนั้น ทำให้เมื่อเจอตัวอักษร 'l' โปรแกรมจะ `break` ออกจากลูปทันที **ก่อน** ที่จะไปถึงบรรทัด `print(char)` ของรอบนั้น ผลคือตัวอักษร 'l' ที่เจอไม่ถูกพิมพ์ออกมาด้วย (เห็นแค่ข้อความ "Found 'l', stopping!")


````python
# Buggy version:
# for char in "hello":
#     if char == 'l':
#         print("Found 'l', stopping!")
#         break
#     print(char)  # This will print 'h', 'e' then break -- 'l' itself is never printed here

# --- Fixed version (print the character first, then check for break) ---
for char in "hello":
    print(char)          # FIX: print the current character first
    if char == 'l':
        print("Found 'l', stopping!")
        break
````

## Activity 3.2 — Pattern Printing with Nested Loops

### 1. Square of Asterisks

````python
size = 5
for row in range(size):
    for col in range(size):
        print("*", end="")
    print()
````

### 2. Right-Angled Triangle

````python
height = 5
for row in range(1, height + 1):
    for col in range(row):
        print("*", end="")
    print()
````

### 3. Challenge: Inverted Right-Angled Triangle

````python
height = 5
for row in range(height, 0, -1):
    for col in range(row):
        print("*", end="")
    print()
````

### 3b. Challenge: Pyramid

````python
height = 5
for row in range(1, height + 1):
    # print leading spaces to center the pyramid
    print(" " * (height - row), end="")
    for col in range(2 * row - 1):
        print("*", end="")
    print()
````
